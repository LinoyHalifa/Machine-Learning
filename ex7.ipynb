```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, train_test_split
from patsy import dmatrix
from statsmodels.api import OLS
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from pygam import LinearGAM, s, f
import matplotlib.pyplot as plt
```

```python


# Load Boston dataset from ISLP
from ISLP import load_data

# Load the Boston dataset
boston = load_data("Boston")

# Extract variables 'dis' and 'nox'
X = boston['dis'].values.reshape(-1, 1)
y = boston['nox'].values

# (a) Fit a cubic polynomial regression
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)
model = LinearRegression().fit(X_poly, y)

# Predict and plot the data
plt.scatter(X, y, color='gray', alpha=0.5, label='Data')
plt.plot(X, model.predict(X_poly), color='red', label='Cubic Polynomial Fit')
plt.xlabel('dis (Distance to employment centers)')
plt.ylabel('nox (Nitrogen Oxides)')
plt.title('Cubic Polynomial Regression')
plt.legend()
plt.show()

# Report regression coefficients
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# (b) Polynomial fits for degrees 1 to 10 and report RSS
rss_values = []
degrees = range(1, 11)
for degree in degrees:
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)
    model = LinearRegression().fit(X_poly, y)
    predictions = model.predict(X_poly)
    rss = np.sum((y - predictions) ** 2)
    rss_values.append(rss)
    
    # Plot each polynomial fit
    plt.plot(X, predictions, label=f'Degree {degree}')

plt.scatter(X, y, color='gray', alpha=0.5, label='Data')
plt.xlabel('dis')
plt.ylabel('nox')
plt.title('Polynomial Fits for Degrees 1-10')
plt.legend()
plt.show()

# Print RSS values
for degree, rss in zip(degrees, rss_values):
    print(f'Degree {degree}: RSS = {rss}')

# (c) Cross-validation for optimal degree
cv_rss = []
for degree in degrees:
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)
    scores = cross_val_score(LinearRegression(), X_poly, y, scoring='neg_mean_squared_error', cv=5)
    cv_rss.append(-scores.mean())

optimal_degree = degrees[np.argmin(cv_rss)]
print("Optimal polynomial degree via cross-validation:", optimal_degree)

# Plot cross-validation results
plt.plot(degrees, cv_rss, marker='o')
plt.xlabel('Polynomial Degree')
plt.ylabel('Cross-Validation MSE')
plt.title('Cross-Validation to Select Optimal Polynomial Degree')
plt.show()

# (d) Fit a regression spline with 4 degrees of freedom
knots = np.percentile(X.flatten(), [25, 50, 75])
spline_basis = dmatrix(f"bs(X, knots={knots.tolist()}, degree=3, include_intercept=True)", {"X": X.flatten()})
spline_model = OLS(y, spline_basis).fit()

# Predict and plot spline fit
plt.scatter(X, y, color='gray', alpha=0.5, label='Data')
plt.plot(X, spline_model.predict(spline_basis), color='blue', label='Regression Spline Fit (4 DoF)')
plt.xlabel('dis')
plt.ylabel('nox')
plt.title('Regression Spline Fit')
plt.legend()
plt.show()

print(spline_model.summary())

# (e) Fit regression splines with varying degrees of freedom
rss_spline = []
spline_degrees = range(3, 11)  # Test degrees of freedom from 3 to 10

for dof in spline_degrees:
    spline_basis = dmatrix(f"bs(X, df={dof}, degree=3, include_intercept=True)", {"X": X.flatten()})
    spline_model = OLS(y, spline_basis).fit()
    predictions = spline_model.predict(spline_basis)
    rss = np.sum((y - predictions) ** 2)
    rss_spline.append(rss)
    
    # Plot each spline fit
    plt.plot(X, predictions, label=f'{dof} DoF')

plt.scatter(X, y, color='gray', alpha=0.5, label='Data')
plt.xlabel('dis')
plt.ylabel('nox')
plt.title('Regression Splines with Varying Degrees of Freedom')
plt.legend()
plt.show()

# Print RSS values for splines
for dof, rss in zip(spline_degrees, rss_spline):
    print(f'Degrees of Freedom {dof}: RSS = {rss}')

# (f) Cross-validation to select best degrees of freedom for splines
cv_rss_spline = []

for dof in spline_degrees:
    spline_basis = dmatrix(f"bs(X, df={dof}, degree=3, include_intercept=True)", {"X": X.flatten()})
    scores = cross_val_score(LinearRegression(), spline_basis, y, scoring='neg_mean_squared_error', cv=5)
    cv_rss_spline.append(-scores.mean())

optimal_dof = spline_degrees[np.argmin(cv_rss_spline)]
print("Optimal degrees of freedom for spline via cross-validation:", optimal_dof)

# Plot cross-validation results for splines
plt.plot(spline_degrees, cv_rss_spline, marker='o')
plt.xlabel('Degrees of Freedom')
plt.ylabel('Cross-Validation MSE')
plt.title('Cross-Validation to Select Optimal Degrees of Freedom for Splines')
plt.show()
```

```python
#(a)

# 1. Load the College dataset
# Ensure the dataset is in the same directory or update the path
college = pd.read_csv('College.csv')  # Replace with the actual dataset location

# 2. Define predictors (X) and response (y)
X = college.drop(columns=['Outstate'])  # Drop the response variable
y = college['Outstate']  # Define the response variable

# 3. Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 4. Perform Forward Stepwise Selection
# Ensure X_train is in numeric format and y_train is 1D
X_train = X_train.select_dtypes(include=[np.number])  # Keep only numeric columns
y_train = y_train.values.ravel()  # Ensure y_train is 1D

# Define a Linear Regression model
model = LinearRegression()

# Initialize the Sequential Feature Selector
sfs = SFS(model,
          k_features='best',  # Automatically find the optimal number of features
          forward=True,  # Perform forward selection
          floating=False,  # Do not allow backward steps
          scoring='r2',  # Use R-squared to evaluate features
          cv=5)  # Use 5-fold cross-validation

# Fit the selector to the training data
sfs = sfs.fit(X_train, y_train)

# Extract the selected features
selected_features = list(sfs.k_feature_names_)
print("Selected features:", selected_features)

# 5. Train the final model using the selected features
X_train_selected = X_train[selected_features]  # Subset X_train with selected features
X_test_selected = X_test[selected_features]    # Subset X_test with the same features

# Fit a statsmodels OLS regression model
final_model = sm.OLS(y_train, sm.add_constant(X_train_selected)).fit()
print(final_model.summary())  # Print the model summary

```

```python
#(b)

# Import necessary libraries
from pygam import LinearGAM, s, f
import matplotlib.pyplot as plt

# 1. Fit GAM model on the selected features
# Initialize a GAM model
gam = LinearGAM(s(0) + s(1) + s(2) + s(3))  # Add smooth terms for each selected feature

# Fit the model on the training data
gam.fit(X_train_selected, y_train)

# 2. Evaluate the model
# Print the model summary
print(gam.summary())

# 3. Plot the results for each predictor
plt.figure(figsize=(15, 8))
for i, feature in enumerate(selected_features):
    plt.subplot(2, 2, i + 1)  # Create a subplot for each feature
    plt.title(f"Partial Dependence of {feature}")
    gam.plot_partial(i)
    plt.xlabel(feature)
    plt.ylabel("Effect on Outstate Tuition")
plt.tight_layout()
plt.show()

# 4. Predict on the test set (optional, for performance evaluation)
y_pred = gam.predict(X_test_selected)

# Calculate Mean Squared Error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error on Test Set: {mse}")
```

```python
# (d)

import matplotlib.pyplot as plt
import math

# Identify terms that are not intercept
non_intercept_terms = [i for i, term in enumerate(gam.terms) if not term.isintercept]

# Number of selected features (ignoring the intercept)
num_features = len(non_intercept_terms)

# Calculate grid size dynamically
num_rows = math.ceil(num_features / 2)  # 2 columns per row
num_cols = 2  # Fixed number of columns

# Plot partial dependence plots for each feature (skip the intercept)
plt.figure(figsize=(15, 5 * num_rows))  # Adjust figure size based on rows
for plot_index, term_index in enumerate(non_intercept_terms):  # Iterate through non-intercept terms
    plt.subplot(num_rows, num_cols, plot_index + 1)  # Dynamically create subplots
    feature_name = selected_features[plot_index]  # Map term index to feature name
    plt.title(f"Partial Dependence for {feature_name}")
    
    # Calculate partial dependence
    XX = gam.generate_X_grid(term=term_index)  # Generate grid for feature
    YY = gam.partial_dependence(term=term_index, X=XX)  # Get partial dependence values
    
    # Plot the partial dependence
    plt.plot(XX[:, term_index], YY, label=feature_name)
    plt.xlabel(feature_name)
    plt.ylabel("Effect on Response")
    plt.legend()
plt.tight_layout()
plt.show()

```

Based on the partial dependence plots, there is no evidence of a non-linear relationship between the predictors (Apps, Accept, Top10perc, Top25perc) and the response variable. All plots show a flat or constant effect, indicating no significant relationship
