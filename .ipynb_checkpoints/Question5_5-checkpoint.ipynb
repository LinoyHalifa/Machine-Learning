```python
# Question A: Fit a logistic regression model that uses income and balance to predict default.


import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split



# Load and preprocess the dataset
default_data = pd.read_csv("Default.csv")
default_data['default'] = default_data['default'].map({'Yes': 1, 'No': 0})
default_data['student'] = default_data['student'].map({'Yes': 1, 'No': 0})


# Fit logistic regression model using the entire dataset
X = default_data[['income', 'balance']]  # Predictors
y = default_data['default']  # Response

logit_model = LogisticRegression()
logit_model.fit(X, y)

# Print model coefficients
print("Intercept (β0):", logit_model.intercept_[0])
print("Coefficient for income (β1):", logit_model.coef_[0][0])
print("Coefficient for balance (β2):", logit_model.coef_[0][1])
```

Question B:
Using the validation set approach, estimate the test error of this
model. In order to do this, you must perform the following steps:
i. Split the sample set into a training set and a validation set.

ii. Fit a multiple logistic regression model using only the training
observations.

iii. Obtain a prediction of default status for each individual in
the validation set by computing the posterior probability of
default for that individual, and classifying the individual to
the default category if the posterior probability is greater
than 0.5.

iv. Compute the validation set error, which is the fraction of
the observations in the validation set that are misclassified.

```python
# Question B: 


# Split data into predictors (income, balance) and response (default)
X = default_data[['income', 'balance']]
y = default_data['default']

# Split into training (50%) and validation (50%) sets
random_seed = 44
#random_seed = 84
#random_seed = 124
X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.5, random_state=random_seed)

# Fit logistic regression model
logit_model = LogisticRegression()
logit_model.fit(X_train, y_train)

# Print coefficients
print("Intercept (β0):", logit_model.intercept_[0])
print("Coefficient for income (β1):", logit_model.coef_[0][0])
print("Coefficient for balance (β2):", logit_model.coef_[0][1])
```

```python
y_prob = logit_model.predict_proba(X_validation)[:, 1]  # Probability for "default = 1"

# Classify as "default" (1) if posterior probability > 0.5
y_pred = (y_prob > 0.5).astype(int)

# Display the results (optional: if you want to see probabilities and classifications)
for i in range(len(y_validation)):
    print(f"True Label: {y_validation.iloc[i]}, Probability of Default: {y_prob[i]:.4f}, Predicted Class: {y_pred[i]}")
```

```python
# Calculate the number of misclassified observations
num_misclassified = (y_validation != y_pred).sum()

# Calculate the total number of observations in the validation set
total_observations = len(y_validation)

# Compute the validation set error
validation_error = num_misclassified / total_observations

print("Validation Set Error:", validation_error)
```

Question D: Now consider a logistic regression model that predicts the probability
of default using income, balance, and a dummy variable
for student. Estimate the test error for this model using the validation
set approach. Comment on whether or not including a
dummy variable for student leads to a reduction in the test error
rate.

```python
# Question D: 

# Add 'student' as a predictor in the model
X = default_data[['income', 'balance', 'student']]  # Include the 'student' variable
y = default_data['default']

# Initialize a list to store validation errors
validation_errors_with_student = []

# Perform validation set approach with seeds 44, 84, 124
for random_seed in [44, 84, 124]:
    # Split the data into training and validation sets
    X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.5, random_state=random_seed)
    
    # Fit logistic regression model
    logit_model = LogisticRegression()
    logit_model.fit(X_train, y_train)
    
    # Predict on the validation set
    y_prob = logit_model.predict_proba(X_validation)[:, 1]  # Probabilities for 'default = 1'
    y_pred = (y_prob > 0.5).astype(int)  # Classify based on threshold of 0.5
    
    # Compute validation error
    num_misclassified = (y_validation != y_pred).sum()
    total_observations = len(y_validation)
    validation_error = num_misclassified / total_observations
    
    # Store the validation error
    validation_errors_with_student.append(validation_error)
    print(f"Random Seed {random_seed} - Validation Error (with student): {validation_error}")


```

Including the student variable increased the validation error across all three splits, suggesting it does not significantly improve the model's performance. This indicates that student may not be a strong predictor for default in this model.
