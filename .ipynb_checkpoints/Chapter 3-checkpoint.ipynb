```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.anova import anova_lm
```

```python
file_path = "Auto.csv"
data = pd.read_csv(file_path, na_values='?')
data_cleaned = data.dropna()
```

```python
#(a)

y = data_cleaned['mpg']
x = data_cleaned['horsepower']
x = sm.add_constant(x)
model = sm.OLS(y, x) # fitting the linear reggration model to the data by OLS
results = model.fit()
print(results.summary())
print(results)




```

i.

by the results - there is a relationship between the "horsepower" variable to the "mpg" variable.
the reasons are:
1. Pvalue is very small - the conclusion is the is a significant relasionship between the two variables
2. the coefficient of the "horsepower" variable is  -0.1578 - the conclusion is negative relationship between the horsepower and the mpg: when the horsepower increse the mpg decrease.
   


ii. 

by the results we can see that the relationship between the "horsepower" and the "mpg" is strong for some reasons:

1. R square is 0.606 and the horsepower variable explain mpg very well.
2. the p- value is signifucant smaller and the meaning is that the predictor ("horsepower") is explaining mpg

iii. 

the relasionship between the predictor ("horsepower") and the response ("mpg") is negative because the coefficient of the "horsepower" variable is negative

```python
# iv. 

intercept = 39.9359
x_parameter = -0.1578
horsepower_value = 98
y= intercept + x_parameter * horsepower_value
print(y)
```

```python
x_new = pd.DataFrame({'mpg': [1] , 'horsepower': [98]})
x_new = sm.add_constant(x_new)  # insert the constant (intercept)

predictions = results.get_prediction(x_new)
pred_summary = predictions.summary_frame(alpha=0.05)
print(pred_summary)
```

the associated 95 %confidence and prediction
intervals are 23.973079 for the lower value and 24.961075 for the upper value

```python
# (b)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Scatter plot of the response (mpg) vs. the predictor (horsepower)
ax.scatter(data_cleaned['horsepower'], data_cleaned['mpg'], label='Data', color='blue')

# Add the least squares regression line using the model's parameters
# The formula for the regression line: mpg = intercept + slope * horsepower
slope = results.params[1] # the slope parameter 
intercept = results.params[0] # the constant (intercept) parameter



xy1 = (0, intercept)  # Starting point (intercept)
xy2 = (1, intercept + slope)  # Second point to define the linear line

# Plot the regression line using axline()
ax.axline(xy1 = xy1, xy2 = xy2, color='red', label='Regression Line')

# Labels and title
ax.set_xlabel('Horsepower')
ax.set_ylabel('MPG')
ax.set_title('Regression of MPG on Horsepower')

# Show legend
ax.legend()

# Display the plot
plt.show()

```

```python
# 1. Residuals vs. Fitted
fitted_values = results.fittedvalues  # the pedicted values
residuals = results.resid  # the residuals from the regrration model results

plt.figure(figsize=(8, 6))
plt.scatter(fitted_values, residuals, color="blue", alpha=0.6)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Fitted Values')
plt.show()
```

```python
# 2. Q-Q Plot
sm.qqplot(residuals, line='45' , fit = True)
plt.title('Normal Q-Q Plot')
plt.show()
```

```python
# 3. Scale-Location Plot
plt.figure(figsize=(8, 6))
sqrt_abs_residuals = np.sqrt(np.abs(residuals))
plt.scatter(fitted_values, sqrt_abs_residuals, color="purple", alpha=0.6)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('|Residuals|')
plt.title('Scale-Location Plot')
plt.show()
```

```python
# 4. Residuals vs. Leverage
sm.graphics.influence_plot(results, criterion="cooks")
plt.title('Residuals vs. Leverage')
plt.show()
```
