```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
file_path = "Auto.csv"
data = pd.read_csv(file_path, na_values='?')
data_cleaned = data.dropna()

```

```python
quant_c = data_cleaned.select_dtypes(include = ['float64', 'int64']).columns
quant_c
```

```python
# Question (a)
qual_c =  data_cleaned.select_dtypes(include = ['object']).columns
# List of quantitative predictors
#quantitative_predictors = ['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year', 'origin']
qual_c
```

```python
# Question (b)

# Calculate and display the range (min and max) for each predictor
for column in quant_c:
    min_value = np.min(data_cleaned[column])
    max_value = np.max(data_cleaned[column])
    range_valu = max_value - min_value
    print(f'Max Value For : {column}:{max_value}')
    print(f'Min Value For : {column}:{min_value}')
    print(f'Range for: {column}: {range_valu}')
    print()
```

```python
# Question (c)

for column in quant_c:
    mean_value = np.mean(data_cleaned[column])
    std_value = np.std(data_cleaned[column])
    print(f"Mean for {column}: {mean_value}") 
    print(f"std for {column}: {std_value}") 
    print()
```

```python
# Question (d)
# Load the data into a DataFrame (replace with your actual file path)


# Remove rows from index 10 to 85
data_subset = data.drop(data.index[9:84])

# Remove rows with any missing values
data_cleaned = data_subset.dropna()

# List of quantitative predictors
quantitative_predictors = ['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year', 'origin']

# Calculate and display the range, mean, and standard deviation for each predictor
for column in quant_c:
    max_value = data_cleaned[column].max()
    min_value = data_cleaned[column].min()
    mean_value = data_cleaned[column].mean()
    std_value = data_cleaned[column].std()
    
    print(f"Range for {column}: {min_value} to {max_value}")
    print(f"Mean : {column}: {mean_value}")
    print(f"Standard deviation for {column}: {std_value}")
    print()
```

```python
#Question (e)
# 1. Scatterplots: Relationship between two predictors (Example: horsepower vs. weight)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_cleaned['horsepower'], y=data_cleaned['weight'], color='blue')
plt.title('Horsepower vs Weight')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.show()

# 2. Pairplot: Relationship between all quantitative predictors
sns.pairplot(data[quant_c])
plt.suptitle("Pairplot of Quantitative Predictors", y=1.02)
plt.show()

# 3. Correlation Heatmap: Visualizing correlation matrix for the predictors
correlation_matrix = data[quant_c].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

findings: strong positive correlation : cylinders between displacement, horsepower between weight
negative correlation : acceleration between horsepower , origin between displacement , origin between weight

```python
#Question(f)

# Scatterplots for continuous variables

continuous_vars = ['displacement', 'horsepower', 'weight', 'acceleration']
for var in continuous_vars:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=data_cleaned, x=var, y='mpg')
    plt.title(f'Relationship between {var} and mpg')
    plt.show()
```

Variables such as weight, horsepower, cylinders, and displacement appear to be particularly significant for predicting mpg. 
These variables show a relatively strong correlation with mpg and can therefore be useful in a prediction model.
