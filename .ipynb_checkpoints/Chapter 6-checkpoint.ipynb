```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from itertools import combinations
from sklearn.linear_model import LassoCV
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LassoCV
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
# Import necessary libraries
from sklearn.linear_model import RidgeCV
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LassoCV
```

```python
# (a)

# Set the seed for reproducibility
np.random.seed(42)

# Parameters
n = 100  # Length of the data

# Generate the predictor X and noise vector epsilon
X = np.random.normal(size=n)
epsilon = np.random.normal(size=n)

# Combine into a DataFrame for analysis
data = pd.DataFrame({'X': X, 'Epsilon': epsilon})

# Display the first few rows of the data
data.head()
```

```python
#(b)

# Set coefficients for the model
beta_0 = 3
beta_1 = 2
beta_2 = -1.5
beta_3 = 0.5

# Generate the response vector Y
Y = beta_0 + beta_1 * X + beta_2 * (X ** 2) + beta_3 * (X ** 3) + epsilon

# Add Y to the DataFrame
data['Y'] = Y

# Display the first few rows of the updated DataFrame
data.head()
```

```python
#(c)

# Create the DataFrame
data = pd.DataFrame({'X1': X, 'Y': Y})  # Rename X to X1 for consistency

# Generate higher-order terms for predictors X2 to X10
for i in range(2, 11):
    data[f'X{i}'] = data['X1'] ** i

# Define the predictors
predictor_names = [f'X{i}' for i in range(1, 11)]

# Calculate the residual variance (sigma^2) from the full model
full_model = LinearRegression()
full_model.fit(data[predictor_names], data['Y'])
y_full_pred = full_model.predict(data[predictor_names])
sigma_squared = mean_squared_error(data['Y'], y_full_pred)

# Perform Forward Stepwise Selection with Cp
selected_predictors = []
Cp_values = []

for step in range(1, len(predictor_names) + 1):  # Forward steps
    best_cp = float("inf")
    best_predictor = None
    
    for predictor in predictor_names:
        if predictor not in selected_predictors:
            # Add this predictor to the current model
            current_predictors = selected_predictors + [predictor]
            X_train = data[current_predictors]
            model = LinearRegression()
            model.fit(X_train, data['Y'])
            y_pred = model.predict(X_train)
            
            # Calculate RSS and Cp
            rss = sum((data['Y'] - y_pred) ** 2)
            p = len(current_predictors) + 1  # Number of predictors + intercept
            cp = rss / sigma_squared + 2 * p - n
            
            if cp < best_cp:
                best_cp = cp
                best_predictor = predictor
    
    # Add the best predictor from this step
    selected_predictors.append(best_predictor)
    Cp_values.append(best_cp)

# Fit the final model with selected predictors
final_model = LinearRegression()
final_model.fit(data[selected_predictors], data['Y'])

# Coefficients of the final model
coefficients = pd.DataFrame({
    "Predictors": ["Intercept"] + selected_predictors,
    "Coefficients": [final_model.intercept_] + list(final_model.coef_)
})

# Display the final selected predictors and coefficients
print("Selected Predictors and Coefficients:")
print(coefficients)

# Display Cp values for each step
print("\nCp values for each step:")
for step, cp in enumerate(Cp_values, start=1):
    print(f"Step {step}: Cp = {cp:.4f}")
```

```python
#(d)


# Backward Stepwise Selection
remaining_predictors = predictor_names[:]  # Start with all predictors
Cp_values_backward = []

for step in range(len(predictor_names), 0, -1):  # Backward steps
    best_cp = float("inf")
    worst_predictor = None

    for predictor in remaining_predictors:
        # Test removing this predictor from the model
        current_predictors = [p for p in remaining_predictors if p != predictor]
        
        # Ensure we have at least one predictor in the model
        if len(current_predictors) == 0:
            continue
        
        X_train = data[current_predictors]
        model = LinearRegression()
        model.fit(X_train, data['Y'])
        y_pred = model.predict(X_train)
        
        # Calculate RSS and Cp
        rss = sum((data['Y'] - y_pred) ** 2)
        p = len(current_predictors) + 1  # Number of predictors + intercept
        cp = rss / sigma_squared + 2 * p - n
        
        if cp < best_cp:
            best_cp = cp
            worst_predictor = predictor
    
    # Remove the worst predictor
    if worst_predictor is not None:
        remaining_predictors.remove(worst_predictor)
        Cp_values_backward.append(best_cp)

# Fit the final model with remaining predictors
if remaining_predictors:  # Ensure there are remaining predictors
    final_model_backward = LinearRegression()  # Corrected the typo
    final_model_backward.fit(data[remaining_predictors], data['Y'])

    # Coefficients of the final model
    coefficients_backward = pd.DataFrame({
        "Predictors": ["Intercept"] + remaining_predictors,
        "Coefficients": [final_model_backward.intercept_] + list(final_model_backward.coef_)
    })

    # Display the final selected predictors and coefficients
    print("Selected Predictors and Coefficients (Backward Stepwise Selection):")
    print(coefficients_backward)

    # Display Cp values for each step
    print("\nCp values for each step (Backward Stepwise Selection):")
    for step, cp in enumerate(Cp_values_backward[::-1], start=1):
        print(f"Step {step}: Cp = {cp:.4f}")
else:
    print("No predictors remaining in the model.")
```

```python
#(e)

X_base = np.random.normal(size=n)
epsilon = np.random.normal(size=n)
beta_0, beta_1, beta_2, beta_3 = 3, 2, -1.5, 0.5
Y = beta_0 + beta_1 * X_base + beta_2 * X_base**2 + beta_3 * X_base**3 + epsilon

# Create the DataFrame
data = pd.DataFrame({'X1': X_base, 'Y': Y})

# Generate higher-order terms for predictors X2 to X10
for i in range(2, 11):
    data[f'X{i}'] = data['X1'] ** i

# Define predictors and response
predictor_names = [f'X{i}' for i in range(1, 11)]
X = data[predictor_names]
y = data['Y']

# Fit Lasso model with cross-validation
lasso = LassoCV(cv=10, random_state=42)  # 10-fold cross-validation
lasso.fit(X, y)

# Optimal value of lambda
optimal_lambda = lasso.alpha_

# Coefficients of the Lasso model
lasso_coefficients = pd.DataFrame({
    "Predictors": predictor_names,
    "Coefficients": lasso.coef_
})

# Plot the cross-validation error as a function of lambda
plt.figure(figsize=(10, 6))
plt.semilogx(lasso.alphas_, lasso.mse_path_.mean(axis=1), label='Mean Cross-Validation Error')
plt.axvline(optimal_lambda, linestyle='--', color='red', label=f'Optimal λ = {optimal_lambda:.4f}')
plt.xlabel('λ (Log Scale)')
plt.ylabel('Mean Cross-Validation Error')
plt.title('Cross-Validation Error vs. λ (Lasso)')
plt.legend()
plt.show()

# Display the resulting coefficients and optimal lambda
print("Optimal λ:", optimal_lambda)
print("Lasso Model Coefficients:")
print(lasso_coefficients)
```

```python
#(a)

# Load the College dataset
# Replace 'College.csv' with the actual path to your dataset
data = pd.read_csv('College.csv')

# Display the first few rows of the dataset to understand its structure
print("First few rows of the dataset:")
print(data.head())

# Split the data into predictors (X) and response variable (y)
# Assuming 'Apps' (number of applications received) is the response variable
X = data.drop(columns=['Apps'])
y = data['Apps']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Display the sizes of the training and test sets
print("\nTraining set size:", X_train.shape[0])
print("Test set size:", X_test.shape[0])
```

```python
#(b)

# Drop unnecessary columns
# Assuming 'Unnamed: 0' is an identifier column that is not needed
data = data.drop(columns=['Unnamed: 0'])

# Convert categorical variable 'Private' to numeric (e.g., Yes=1, No=0)
data['Private'] = data['Private'].apply(lambda x: 1 if x == 'Yes' else 0)

# Split the data into predictors (X) and response variable (y)
X = data.drop(columns=['Apps'])  # Remove the target variable
y = data['Apps']  # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Fit a linear model using least squares on the training set
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate the test error (Mean Squared Error)
test_error = mean_squared_error(y_test, y_pred)

# Display the test error
print(f"Test Error (Mean Squared Error): {test_error:.4f}")
```

```python
#(c)

# Define a range of λ (alphas) to test
alphas = [0.1, 1, 10, 100, 1000]

# Fit Ridge Regression with Cross-Validation
ridge_model = RidgeCV(alphas=alphas, cv=10)  # 10-fold Cross-Validation
ridge_model.fit(X_train, y_train)

# Optimal λ (alpha)
optimal_lambda = ridge_model.alpha_

# Make predictions on the test set
y_pred_ridge = ridge_model.predict(X_test)

# Calculate the test error (Mean Squared Error)
test_error_ridge = mean_squared_error(y_test, y_pred_ridge)

# Display the results
print(f"Optimal λ (alpha) chosen by Cross-Validation: {optimal_lambda:.4f}")
print(f"Test Error (Mean Squared Error) for Ridge Regression: {test_error_ridge:.4f}")
```

```python
#(d)

# Define a range of λ (alphas) to test
alphas = np.logspace(-3, 3, 100)  # Test 100 values between 0.001 and 1000

# Fit Lasso Regression with Cross-Validation
lasso_model = LassoCV(alphas=alphas, cv=10, random_state=42)  # 10-fold Cross-Validation
lasso_model.fit(X_train, y_train)

# Optimal λ (alpha)
optimal_lambda_lasso = lasso_model.alpha_

# Make predictions on the test set
y_pred_lasso = lasso_model.predict(X_test)

# Calculate the test error (Mean Squared Error)
test_error_lasso = mean_squared_error(y_test, y_pred_lasso)

# Count the number of non-zero coefficients
non_zero_coefficients = np.sum(lasso_model.coef_ != 0)

# Display the results
print(f"Optimal λ (alpha) chosen by Cross-Validation: {optimal_lambda_lasso:.4f}")
print(f"Test Error (Mean Squared Error) for Lasso Regression: {test_error_lasso:.4f}")
print(f"Number of non-zero coefficients: {non_zero_coefficients}")
```
