```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "Boston.csv"
df = pd.read_csv(file_path, na_values='?')
```

```python
# Question (a)
mat_dim = df.shape
mat_dim
```

```python
#Question (c)
selected_predictors = ["crim", "zn", "indus", "chas"]
predictors = df[selected_predictors]
sns.pairplot(predictors)
```

CRIM VS ZN - It seems that in certain areas, crime increases when the percentage of residential land area stay low. when the percentage of residential land area increases, crime rates in certain areas are not change and stay low.

CRIM vs INDUS - It seems that an increase in the percentage of land designated for industrial use does not raise the crime rate, except for exceptional cases where there is an increase in the crime rate in areas with a high percentage of land designated for industrial use.

CRIM vs CHAS - It seems that places not that not bordering the Charles River have a higher crime rate, while places that border the Charles River have a lower crime rate

ZN VS INDUS- It seems that in some places, when the percentage of land designated for industrial use is low, the percentage of land designated for residential use is high. However, after a certain point where the percentage of land designated for industrial use increases, the percentage of land designated for residential use is not change and stay low.

ZN VS CHAS - In areas that are not bordering to the river, the percentage of large plots is generally low with a wide spread. In contrast, in areas thas bordering to the river, there are fewer samples, but some show higher ZN percentages, which may indicate more upscale areas.

INDUS VS CHAS - It seems that most observations show that in areas not bordering the river, there are more areas with land designated for industrial use. In contrast, in areas bordering the river, there are fewer observations, but some show that in these areas there are industrial land uses, while in other areas there is less industrial land use.

```python
#Question (e)
print(df.describe())
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.histplot(df['crim'], kde=True, ax=axes[0])
axes[0].set_title("Crime Rate Graph (CRIM) ")

sns.histplot(df['tax'], kde=True, ax=axes[1])
axes[1].set_title("Taxes Rate Graph (TAX)")

sns.histplot(df['ptratio'], kde=True, ax=axes[2])
axes[2].set_title("PTratio Rate Graph (PTRATIO)")


plt.tight_layout()
plt.show()

# high CRIM rates
high_crim = df['crim'] > df['crim'].max() * 1.5
low_crim = df['crim'] < df['crim'].min() * 0.5 

# high TAX rates
high_tax = df['tax'] > df['tax'].max() * 1.5  
low_tax = df['tax'] < df['tax'].min() * 0.5 

# high PTRATIO rates
high_ptratio = df['ptratio'] > df['ptratio'].max() * 1.5  
low_ptratio = df['ptratio'] < df['ptratio'].min() * 0.5 


print(f'Hight Crime rate : {high_crim.head()}')  
print(f'low Crime rate : {low_crim.head()}')
print('-' * 20)
print()
print(f'Hight Tax rate : {high_tax.head()}')  
print(f'low Tax rate : {low_tax.head()}')
print('-' * 20)
print()
print(f'Hight Ptratio rate : {high_ptratio.head()}')  
print(f'low Ptratio rate : {low_ptratio.head()}')
```

```python
#Question (f)
charles_river_bound = df[df['chas'] == 1]
num_charles_river_suburbs = len(charles_river_bound)
print(f"Number of suburbs that bound the Charles River: {num_charles_river_suburbs}")
```

```python
# Question (g)

ptratio_median = df['ptratio'].median()
print(f'The Median Of PTRATIO is : {ptratio_median}')
```

```python
# Question (h)

lowest_medv_suburb = df.loc[df['medv'].idxmin()]
lowest_medv_values = lowest_medv_suburb.drop('medv')
summary_stats = df.describe()

print(f'The City With The Lowest MEDV Value:\n{lowest_medv_suburb}')
print('-'*40)
print()
print(f'The General Statistics Of the Predictors:\n{summary_stats}')
print('-'*40)
print()
print(f'The Predictors Values Of The Lowest MEDV:\n{lowest_medv_values}')
print('-'*40)
print()
```

The data indicates that the suburb with the lowest MEDV (5.000) is an area with a high crime rate, extensive industrial areas, high air pollution, and a low amount of residential land. Additionally, it has very small apartments, very old buildings, and an education system with a high student-to-teacher ratio. All of these factors suggest that this is an area with a low home value, and it seems to have a population with a low socioeconomic status.
