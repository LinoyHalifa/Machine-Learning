```python
y_pred_prob = result.predict(X)

# Classify predictions as 1 ("Up") if probability > 0.5, else 0 ("Down")
y_pred = [1 if prob > 0.5 else 0 for prob in y_pred_prob]
conf_matrix = confusion_matrix(y, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Compute accuracy
accuracy = accuracy_score(y, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

explaination: 

54: True Negatives (TN) - the model predicted 54 times "down" correctly when the real result was "down"
557 : True Positive (TP)  - the model predicted 557 times "up" correctly when the real result was "up"
430 : False Positive (FP) -the model predicted 430 times "up" incorrectly when the real result was "down"
48 : False Positive (FN) -the model predicted 48 times "down" incorrectly when the real result was "up"

the total true predictions : 
557 + 54 = 611

the total false predictions : 

430 + 48 = 478.


the accuracy of the model :

TP + TN / total observation  = 0.56

```python
X_train_full = train_data[['Lag1', 'Lag2', 'Lag3', 'Lag4', 'Lag5', 'Volume']]
y_train = train_data['DirectionBinary']
X_test_full = test_data[['Lag1', 'Lag2', 'Lag3', 'Lag4', 'Lag5', 'Volume']]
y_test = test_data['DirectionBinary']

# test function
def test_methods(X_train, y_train, X_test, y_test):
    methods = {
        "Logistic Regression": LogisticRegression(),
        "LDA": LinearDiscriminantAnalysis(),
        "QDA": QuadraticDiscriminantAnalysis(),
        "Naive Bayes": GaussianNB(),
    }

    results = {}

    for method_name, model in methods.items():
        # model fitting
        model.fit(X_train, y_train)
        # חיזוי
        y_pred = model.predict(X_test)
        # confusion matrix caculation and accuracy
        conf_matrix = confusion_matrix(y_test, y_pred)
        accuracy = accuracy_score(y_test, y_pred)
        results[method_name] = (conf_matrix, accuracy)

    # KNN with diffrent values of k
    for k in [1, 3, 5, 7]:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)
        conf_matrix = confusion_matrix(y_test, y_pred)
        accuracy = accuracy_score(y_test, y_pred)
        results[f"KNN (K={k})"] = (conf_matrix, accuracy)

    return results

# checking the all posobilities
best_result = {"Method": None, "Variables": None, "Confusion Matrix": None, "Accuracy": 0}

for num_vars in range(1, X_train_full.shape[1] + 1):
    for vars_subset in itertools.combinations(X_train_full.columns, num_vars):
        X_train = train_data[list(vars_subset)]
        X_test = test_data[list(vars_subset)]
        results = test_methods(X_train, y_train, X_test, y_test)

        for method_name, (conf_matrix, accuracy) in results.items():
            if accuracy > best_result["Accuracy"]:
                best_result["Method"] = method_name
                best_result["Variables"] = vars_subset
                best_result["Confusion Matrix"] = conf_matrix
                best_result["Accuracy"] = accuracy

# best result presentation
print("Best Method:", best_result["Method"])
print("Best Variables:", best_result["Variables"])
print("Confusion Matrix:\n", best_result["Confusion Matrix"])
print(f"Accuracy: {best_result['Accuracy']:.2f}")
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_selected)
X_test_scaled = scaler.transform(X_test_selected)

# k list value for chacking
k_values = [1, 3, 5, 7, 9]
test_errors = []

# running k values for KNN model    
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    y_pred = knn.predict(X_test_scaled)
    
    # error test calculation
    accuracy = accuracy_score(y_test, y_pred)
    test_error = 1 - accuracy
    test_errors.append(test_error)
    
    print(f"K = {k}: Test Error = {test_error:.2f}")

# identifying the best value of K
best_k = k_values[test_errors.index(min(test_errors))]
print(f"\nBest K: {best_k} with Test Error = {min(test_errors):.2f}")
```
