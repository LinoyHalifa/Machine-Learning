```python
#(a)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
import numpy as np



rng = np.random.default_rng(5)  
# Create a random number generator with a fixed seed for reproducibility

x1 = rng.uniform(size=500) - 0.5  
# Generate 500 random samples from a uniform distribution and shift them to the range [-0.5, 0.5]

x2 = rng.uniform(size=500) - 0.5  
# Generate another 500 random samples for the second variable, also in the range [-0.5, 0.5]

y = x1**2 - x2**2 > 0  
# Define the class labels: True if x1 squared is greater than x2 squared, False otherwise
```

```python
# (b)


# Plot the observations with different colors based on their class labels
plt.figure(figsize=(8, 6))
plt.scatter(x1[y], x2[y], color='blue', label='Class 1')  
# Points where x1^2 > x2^2, labeled as Class 1 (blue)

plt.scatter(x1[~y], x2[~y], color='red', label='Class 2')  
# Points where x1^2 <= x2^2, labeled as Class 2 (red)

plt.xlabel('$X_1$')  
# Label for the x-axis

plt.ylabel('$X_2$')  
# Label for the y-axis

plt.title('Scatter Plot of Observations Colored by Class Labels')  
# Title of the plot

plt.legend()  
# Display the legend to distinguish between the classes

plt.show()
```

```python
#(c)

# Stack x1 and x2 into a single feature matrix
X = np.column_stack((x1, x2))  
# Combine x1 and x2 into a matrix with two columns (features)

# Convert boolean labels to integers (0 and 1)
y_int = y.astype(int)  
# Convert True/False to 1/0 for logistic regression

# Initialize the logistic regression model
log_reg = LogisticRegression()  
# Create an instance of the logistic regression model

# Fit the model to the data
log_reg.fit(X, y_int)  
# Train the model using the features X and labels y_int
```

```python
#(d)

# Predict the class labels for the training data
y_pred = log_reg.predict(X)  
# Use the trained logistic regression model to predict class labels

# Plot the observations colored by predicted class labels
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

# Plot observations predicted as Class 1
plt.scatter(x1[y_pred == 1], x2[y_pred == 1], color='blue', label='Predicted Class 1')  
# Points predicted as Class 1 (blue)

# Plot observations predicted as Class 0
plt.scatter(x1[y_pred == 0], x2[y_pred == 0], color='red', label='Predicted Class 0')  
# Points predicted as Class 0 (red)

plt.xlabel('$X_1$')  
# Label for the x-axis

plt.ylabel('$X_2$')  
# Label for the y-axis

plt.title('Predicted Class Labels with Linear Decision Boundary')  
# Title of the plot

plt.legend()  
# Display the legend to distinguish predicted classes

plt.show()
```

```python
# (e)

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression

# Create polynomial features (including squares and interaction terms)
poly = PolynomialFeatures(degree=2, include_bias=False)  
# Generate polynomial features up to degree 2 without adding a bias term

X_poly = poly.fit_transform(X)  
# Transform the original features into polynomial features

# Initialize the logistic regression model
log_reg_poly = LogisticRegression()  
# Create an instance of the logistic regression model

# Fit the model using the non-linear features
log_reg_poly.fit(X_poly, y_int)  
# Train the model using the polynomial features and labels
```

```python
#(f)

# Predict the class labels using the non-linear logistic regression model
y_pred_poly = log_reg_poly.predict(X_poly)  
# Predict class labels using the polynomial features

# Plot the observations colored by predicted class labels (non-linear model)
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

# Plot points predicted as Class 1
plt.scatter(x1[y_pred_poly == 1], x2[y_pred_poly == 1], color='blue', label='Predicted Class 1')  
# Points predicted as Class 1 (blue)

# Plot points predicted as Class 0
plt.scatter(x1[y_pred_poly == 0], x2[y_pred_poly == 0], color='red', label='Predicted Class 0')  
# Points predicted as Class 0 (red)

plt.xlabel('$X_1$')  
# Label for the x-axis

plt.ylabel('$X_2$')  
# Label for the y-axis

plt.title('Predicted Class Labels with Non-Linear Decision Boundary')  
# Title of the plot

plt.legend()  
# Display the legend to distinguish predicted classes

plt.show()
```

```python
#(a)

import pandas as pd

# Load the Auto dataset
file_path = "Auto.csv"
data = pd.read_csv(file_path, na_values='?')
auto_data = data.dropna()

# Check the first few rows to understand the structure of the data
auto_data.head()  
# Display the first few rows of the dataset to inspect the columns

# Calculate the median gas mileage (assuming 'mpg' is the gas mileage column)
median_mpg = auto_data['mpg'].median()  
# Compute the median of the 'mpg' (miles per gallon) column

# Create a binary variable: 1 if mpg > median, 0 otherwise
auto_data['high_mileage'] = (auto_data['mpg'] > median_mpg).astype(int)  
# Create a new column 'high_mileage' where cars with mpg above the median are labeled 1, others 0

# Display the updated dataset with the new binary variable
auto_data[['mpg', 'high_mileage']].head()  
# Show the 'mpg' and 'high_mileage' columns to verify the transformation

```

```python
#(b)

from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
import numpy as np

# Prepare the data by removing 'mpg' and any non-numeric columns
features = auto_data.drop(['mpg', 'high_mileage', 'name'], axis=1)  
# Remove 'mpg' (target variable), 'high_mileage' (binary target), and 'name' (non-numeric column)

# Handle missing values if any
features = features.dropna()  
# Drop rows with missing values (if they exist)

# Target variable
target = auto_data.loc[features.index, 'high_mileage']  
# Ensure target matches the cleaned feature dataset

# Define a range of C values to test
C_values = [0.01, 0.1, 1, 10, 100]  
# Test various regularization strengths

# Store cross-validation errors
cv_errors = []

# Perform cross-validation for each C value
for C in C_values:
    svc = SVC(C=C, kernel='linear')  
    # Initialize the SVM classifier with a linear kernel and current C value
    
    scores = cross_val_score(svc, features, target, cv=5, scoring='accuracy')  
    # Perform 5-fold cross-validation and compute accuracy scores
    
    error = 1 - scores.mean()  
    # Calculate the cross-validation error (1 - accuracy)
    
    cv_errors.append(error)  
    # Store the error for analysis

# Display the cross-validation errors for each C value
for C, error in zip(C_values, cv_errors):
    print(f'C={C}: Cross-Validation Error = {error:.4f}')
```

```python
#(c)




# Define parameter grids for RBF and Polynomial kernels
C_values = [0.1, 1, 10]
gamma_values = [0.01, 0.1, 1]
degrees = [2, 3, 4]

# Store results for RBF kernel
rbf_results = []

# Perform cross-validation for RBF kernel
for C in C_values:
    for gamma in gamma_values:
        svc_rbf = SVC(kernel='rbf', C=C, gamma=gamma)
        scores = cross_val_score(svc_rbf, features, target, cv=5, scoring='accuracy')
        error = 1 - scores.mean()
        rbf_results.append((C, gamma, error))

# Display RBF kernel results
print("RBF Kernel Results:")
for C, gamma, error in rbf_results:
    print(f'C={C}, gamma={gamma}: Cross-Validation Error = {error:.4f}')


# Store results for Polynomial kernel
poly_results = []

# Perform cross-validation for Polynomial kernel
for C in C_values:
    for degree in degrees:
        svc_poly = SVC(kernel='poly', C=C, degree=degree)
        scores = cross_val_score(svc_poly, features, target, cv=5, scoring='accuracy')
        error = 1 - scores.mean()
        poly_results.append((C, degree, error))

# Display Polynomial kernel results
print("\nPolynomial Kernel Results:")
for C, degree, error in poly_results:
    print(f'C={C}, degree={degree}: Cross-Validation Error = {error:.4f}')
```
